```python
import pandas as pd
import numpy as np

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25.0, 27.6, 30.0, 29.6, 28.0],
    'City': ['Cairo', 'Paris', 'Paris', 'London', 'Paris'],
    'Math Score': [90, 75, 85, 70, 88],
    'Science Score': [95, 65, 82, 60, 91]
}
df = pd.DataFrame(data)





```

```python
df['Math Score']
```

```python
df[df['Science Score'] > 80]
```

```python
df['Average']=(df['Science Score']+df['Math Score'])/2
df
```

```python
df.sort_values('Average', ascending=True)
```

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, np.nan, 30, np.nan, 28],
    'City': ['Cairo', 'Paris', None, 'London', 'Paris'],
    'Salary': [5000, 6000, None, 5500, 5800]
}
df2=pd.DataFrame(data)
df2
```

```python
df2.isnull().sum()
```

```python
df2['Age']=df2['Age'].fillna(df['Age'].mean())
df2
```

```python
df2['City']=df2['City'].fillna('Unknown')
df2
```

```python
df2=df2.dropna(subset=['Salary'])
df2
```
